```python
train_data =[]
train_label = []
train_data_en =[]
train_data_fr =[]
train_data_in =[]
train_label_en = []
train_label_fr = []
train_label_in = []
```

```python
files = glob.glob("./train/*.txt")
for file_name in files:
    # Get Lables
    basename = os.path.basename(file_name)
    language = basename.split("_")[0]
    # Export text
    file = open(file_name, "r", encoding="utf-8")
    text = file.read()
    text = text.lower()
    file.close
    # Get frequency of each letter appearance
    code_a = ord('a')
    code_z = ord('z')
    count = [0 for n in range(0,26)]
    for letter in text:
        code_x = ord(letter)
        if code_a <= code_x <= code_z:
            count[code_x - code_a] += 1
    # Normalizing
    total = sum(count)
    freq = list(map(lambda n: n / total, count))
    # Insert into created list
    train_data.append(freq)
    train_label.append(language)
    if language == "en":
        train_data_en.append(freq)
        train_label_en.append(language)
    elif language == "fr":
        train_data_fr.append(freq)
        train_label_fr.append(language)
    elif language == "in":
        train_data_in.append(freq)
        train_label_in.append(language)
    else :
        continue
    
```

```python
test_data =[]
test_label = []

test_data_en =[]
test_data_fr =[]
test_data_in =[]
test_label_en = []
test_label_fr = []
test_label_in = []
```

```python
files = glob.glob("./test/*.txt")
for file_name in files:
    # Get Lables
    basename = os.path.basename(file_name)
    language = basename.split("-")[0]
    
    
    # Export text
    file = open(file_name, "r", encoding="utf-8")
    text = file.read()
    text = text.lower()
    file.close
    # Get frequency of each letter appearance
    code_a = ord('a')
    code_z = ord('z')
    count = [0 for n in range(0,26)]
    for letter in text:
        code_x = ord(letter)
        if code_a <= code_x <= code_z:
            count[code_x - code_a] += 1
    # Normalizing
    total = sum(count)
    freq = list(map(lambda n: n / total, count))
    # Insert into created list
    test_data.append(freq)
    test_label.append(language)
    if language == "en":
        test_data_en.append(freq)
        test_label_en.append(language)
    elif language == "fr":
        test_data_fr.append(freq)
        test_label_fr.append(language)
    elif language == "in":
        test_data_in.append(freq)
        test_label_in.append(language)
    else :
        continue
```

```python

clf = svm.SVC()
start = time.time()
# traning

clf.fit(train_data, train_label)
end = time.time()
print(end - start)
# Predict
svc_predict = clf.predict(test_data)
```
